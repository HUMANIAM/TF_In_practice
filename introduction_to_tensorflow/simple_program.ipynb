```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
xs = np.arange(-1, 5, dtype="float64")
ys = 2*xs - 1
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```

### Steps to create a model.
----------------------------
1- `model representation` (set layers, input and output of every layer)  
2- `compile the model` (by specify the optimizer algo and the loss function)  
3- `fit the model` (set the actual input, the actual output, the epochs)   

### Load/Save the model.
`source:` [save_and_load](https://www.tensorflow.org/tutorials/keras/save_and_load)

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
print("train shape : ", train_images.shape )
print("test shape : ", test_images.shape )
print("train_labels: ", train_labels.shape)
```

```python
# simple preprocessing on the input image.
train_labels = train_labels[:1000]
test_labels = test_labels[:1000]

train_images = train_images[:1000].reshape(-1, 28 * 28) / 255.0
test_images = test_images[:1000].reshape(-1, 28 * 28) / 255.0
```

```python
def create_seq_model(in_shape):
    # 1- model representation.
    model = keras.models.Sequential([
        keras.layers.Dense(units=512, activation='relu', input_shape=(in_shape,)),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(units=10)
    ])
    
    # 2- compile the model. set the optimizer and the loss function.
    model.compile( optimizer="adam", 
                   loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
                   metrics=[tf.metrics.SparseCategoricalAccuracy()]
                 )
    return model
    
```

```python
model = create_seq_model(train_images.shape[1])
```
